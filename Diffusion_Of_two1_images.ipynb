```python
import os
from PIL import Image
```

```python
shirtImage = Image.open('/content/Cloth Image.jpg')
modelImage = Image.open('/content/Model Image.png')

size = (min(shirtImage.size[0], modelImage.size[0]),
        min(shirtImage.size[1], modelImage.size[1]))

shirtImage_resized = shirtImage.resize(size)
modelImage_resized = modelImage.resize(size)

shirtImage_rgb = shirtImage_resized.convert("RGB")
modelImage_rgb = modelImage_resized.convert("RGB")
```

```python
# !pip install git+https://github.com/Gothos/diffusers.git@flux-inpaint
```

```python
import torch
from diffusers import FluxImg2ImgPipeline
from diffusers.utils import make_image_grid

# Load the Flux diffusion pipeline
device = "cuda" 
pipe = FluxImg2ImgPipeline.from_pretrained("black-forest-labs/FLUX.1-schnell", torch_dtype=torch.float32)
pipe = pipe.to(device)
```

```python
# Define the prompt
prompt = "A realistic image of a male model wearing a shirt from a different image. The shirt should be naturally integrated."

# Perform the image transformation
result_image = pipe(prompt=prompt, init_image=modelImage_rgb, image=shirtImage_rgb).images[0]
```

```python
# Create a grid of images
grid_image = make_image_grid([model_image, shirt_image, result_image], rows=1, cols=3)

# Display the grid image
grid_image.show()
```
